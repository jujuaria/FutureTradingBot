# Download historical data

```python
import matplotlib.pyplot as plt
from arch import arch_model
import pandas_ta
import pandas as pd
import numpy as np
import os
import yfinance as yf
from datetime import datetime, date, timedelta

%matplotlib inline
```

```python
end_date = datetime.today().strftime('%Y-%m-%d')
start_date = datetime.today() - timedelta(days=365)

futures_symbol = "MES=F"
stock_symbol = "SPY"
```

```python
# Download the platinum prices
futures_data = yf.download(futures_symbol, start=start_date, end=end_date)
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)
# Set the index to a datetime type
futures_data.index = pd.to_datetime(futures_data.index)
stock_data.index = pd.to_datetime(stock_data.index)
```

```python
# code 
# Visualizing The Open Price of all the stocks 

# to set the plot size 
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(futures_data.index, futures_data["Adj Close"], 'g-')
ax2.plot(stock_data.index, stock_data["Adj Close"], 'b-')

ax1.set_xlabel('Time')
ax1.set_ylabel('futures data', color='y')
ax2.set_ylabel('stock data', color='b')

plt.show()
```
